```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
tags = table.find_all('td', class_='no-alt', attrs={'style':'white-space:nowrap;'})
labels = []

d = dict()

for tag in tags:
    print(tag.get_text())
    labels.append(tag.get_text())
    d[tag.get_text()] = dict()
```

```python
import re

re_date = re.compile('2019/0[1-9]|2019/1[0-2]')
re_val = re.compile('[0-9.]*')

date = ''
value = ''
rows = table.find_all('td')

tag_id = 0
count = 0
found_pair = False

for row in rows:
    text = row.get_text()
    date_match = re_date.search(text)
    val_match = re_val.search(text)
    if date_match:
        result = date_match.group()
        date = result
    elif val_match:
        result = val_match.group()
        if result != '':
            value = result
            found_pair = True
    if found_pair:
        found_pair = False
        tag = labels[tag_id]
        d[tag][date] = float(value)
        count += 1
        if count == 11:
            tag_id += 1
            count = 0
```
