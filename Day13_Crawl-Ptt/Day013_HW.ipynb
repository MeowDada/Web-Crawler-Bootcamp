## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'

r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
```

```python
print(soup.body)
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
article = soup.find(class_='r-ent')
title = article.find(class_='title').a.text
author = article.find(class_='meta').find(class_='author').text
date = article.find(class_='meta').find(class_='date').text
print('作者:', author)
print('標題:', title)
print('時間:', date)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
articles = soup.find_all(class_='r-ent')
for article in articles:
    try:
        title = article.find(class_='title').a.text
        author = article.find(class_='meta').find(class_='author').text
        date = article.find(class_='meta').find(class_='date').text
        print('{}.{}.{}'.format(author, title, date))
    except:
        pass
```
