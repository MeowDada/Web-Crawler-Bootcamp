```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.html'

r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
```

```python
print(soup)
```

```python
for d in soup.find(class_='part_list_2').find_all('h3'):
    date = d.find(class_='date').text
    tag = d.find(class_='tag').text
    title = d.find('a').text
    print(date, tag, title)
```

### ① 取出今天所有的發文

```python
from datetime import date

today = date.today()
today_datetime = today.strftime("%Y/%m/%d")
print(today_datetime)
```

```python
for d in soup.find(class_='part_list_2').find_all('h3'):
    date = d.find(class_='date').text
    if date[:10] == today_datetime:
        tag = d.find(class_='tag').text
        title = d.find('a').text
        print(date, tag, title)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
news = {}

for d in soup.find(class_='part_list_2').find_all('h3'):
    date = d.find(class_='date').text
    if date[:10] == today_datetime:
        tag = d.find(class_='tag').text
        title = d.find('a').text
        if tag in news:
            news[tag].append(title)
        else:
            news[tag] = [title]

for k, v in news.items():
    for title in v:
        print(k, title)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
max_count = 0
max_tag = None

for k, v in news.items():
    if len(v) > max_count:
        max_count = len(v)
        max_tag = k
print(max_tag)
```
